### Question 1: Recreate the following plot using Matplotlib’s functional API. This line plot shows the populationof Canada between the years 1955 and 2020. Make sure to include the axes labels, title, and thatthe tick marks on the y-axis match the figure below.

```python
# Creating a df with only value of country is Canada
canada_df = df.loc[df['Country'] == 'Canada']
canada_df
```

### Question 2 :Recreate the following plot using Matplotlib’s functional API. This line plot shows the percentage of the world’s population that are living in urban areas. This has increased from low 30% to mid50% over the past 60 years.

```python
# Creating a population df which is grouped by Year and consist of sum of  population  , sum of urban population for each Year
pop_df=df[['Population','UrbanPopulation','Year']].groupby('Year').sum()
# Calculating  Urban Percentage
pop_df['Percentage']=(pop_df['UrbanPopulation']/pop_df['Population'])*100
pop_df
```

```python
# Creating a plot using Population df  Year Vs Urban Percentage
fig,ax = plt.subplots(figsize=(8, 6))
x = pop_df.index
y = pop_df['Percentage']
ax.plot(x,y)
#Setting Titles and labels
plt.xlabel('Year')
plt.ylabel('Urban Percentage')
plt.title("Percentage of Global Population living in Urban Areas")
plt.show()
# For Changing Labels in Y axis
ytick_labels = ["30%","35%","40%","45%","50%","55%"]
ax.set_yticklabels(ytick_labels)
plt.show()
```

### Question 3: Recreate the following plot using Matplotlib’s object-oriented API. This line plot shows the yearly population change (absolute number) per continent over the past 60 years. You will need to compute this data on a per continent basis using the country values. Make sure to include the axes labels and plot title, and that the colours of each line match the figure. The legend should be in the centre of the plot without a border (hint: framealpha), and the continents in alphabetical order.

```python
# Creating df grouped by Continents with aggregating Yearly Population Change
continent_pop_df = df.groupby(['Continent','Year'], as_index=False).agg({'YearlyChange':'sum'}).sort_values('Year', ascending=True)
continent_pop_df = continent_pop_df.set_index('Continent')
continent_pop_df
```

```python
# Creating a function to plot
def create_lineplot(df):
    fig = plt.Figure()
    fig,ax = plt.subplots(figsize=(8, 6))
    ax = fig.add_axes([0, 0, 1, 1])
    # Plotting using  Year and Yearly Change in df grouped by continent 
    # Colors may be slightly inaccurate since it is picked used Online Color Picker.
    ax.plot(df[df.index == 'Africa']['Year'], df[df.index == 'Africa']['YearlyChange'],color='#1f77b4',label="Africa")
    ax.plot(df[df.index == 'Asia']['Year'], df[df.index == 'Asia']['YearlyChange'],color='#ff7f0f',label="Asia")
    ax.plot(df[df.index == 'Europe']['Year'], df[df.index == 'Europe']['YearlyChange'],color='#2ba02b',label="Europe")
    ax.plot(df[df.index == 'Latin America and the Caribbean']['Year'], df[df.index == 'Latin America and the Caribbean']['YearlyChange'],color='#d52828',label="Latin America and the Caribbean")
    ax.plot(df[df.index == 'North America']['Year'], df[df.index == 'North America']['YearlyChange'],color='#9267bd',label="North America")
    ax.plot(df[df.index == 'Oceania']['Year'], df[df.index == 'Oceania']['YearlyChange'],color='#8c564c',label="Oceania")
    #Setting Titles and labels
    ax.set_title('Yearly Population Change by Continent')
    ax.set_xlabel('Year')
    ax.set_ylabel('Yearly Population Change')
    ax.legend(loc ='center', framealpha = False)
    return fig,ax
    
fig,ax=create_lineplot(continent_pop_df)
# Source Attribution
# This Code Inspired from Similar kind of code in Last Sem Data Preparation Project(Creating a Function for Plotting Multiple line plots in a fig)

```

### Question 4: Recreate the following plot using Matplotlib’s object-oriented API. This scatter plot shows the median age and percentage population growth (Yearly%Change) from the year 2019 to 2020 for all the countries of the world. Each country is colour coded according to its continent. Notice how countries with younger median ages have larger population growth rates. Make sure to include the axes labels and plot title, and that the colours match the figure. (The colours are consistent with the plot from question 3 – this reinforces their interpretation.) Exclude countries that are missing median age information. The legend should be in the lower left corner, without a border, and the continents in alphabetical order.

```python
#Creating the df to plot the necessary graph
median_df = df[(df.Year == 2020) & (df.MedianAge > 15)]
median_df= median_df.sort_values(by=['Continent'])
median_df
```

```python
#plotting the scatter plot  using median df MedianAge Vs Percentage
fig,ax = plt.subplots(figsize=(8, 6))
for name,data in median_df.groupby('Continent'):
    ax.scatter(data['MedianAge'], data['Yearly%Change'], label=name)
plt.legend(loc ='lower left',framealpha=False)
#Setting Titles and labels
ax.set_title('Country Population Growths (2019 to 2020)')
ax.set_xlabel('Median Age')
ax.set_ylabel('Percent')
ax.plot()
plt.show()

#Source Attribution
# This Code Inspired from the below Stack Overflow page
#https://stackoverflow.com/questions/21654635/scatter-plots-in-pandas-pyplot-how-to-plot-by-category
```

### Question 5: Recreate the following plot using Matplotlib’s object-oriented API. This bar plot shows the population of the top 10 most populous countries in 2020.Make sure to include the axes labels and title, and that the colours and x-axis ticks match the figure.The countries should be in descending order in according to their population.

```python
#Creating a df grouping by Country and adding Population in 2020 and sort by descending
pop_top10_df = df[(df.Year == 2020)].groupby('Country',as_index=False).agg({'Population':'sum'}).sort_values('Population', ascending=False)
pop_top10_df = pop_top10_df.head(10)
pop_top10_df
```

```python
#Plotting the Bar chart using pop top10 df - horizontally using barh
fig, axs = plt.subplots(figsize = (8,6))
bars = plt.barh(pop_top10_df["Country"], pop_top10_df["Population"])
# Setting color for each Country
# Colors may be slightly inaccurate since it is picked using Online Color Picker.
bars[0].set_color('#2276b5')
bars[1].set_color('#eb8333')
bars[2].set_color('#008100')
bars[3].set_color('#ff0000')
bars[4].set_color('#a75cf2')
bars[5].set_color('#5f4034')
bars[6].set_color('#e080db')
bars[7].set_color('#808080')
bars[8].set_color('#9bac54')
bars[9].set_color('#30bfc0')
#Setting Titles and labels
axs.set_title("Top 10 Countries by Population (2020)")
xticks_labels = ["0", "200M", "400M", "600M", "800M","1B", "1.2B","1.4B"]
axs.set_xticklabels(xticks_labels)
# To Show Bar Graph in descending order
axs.invert_yaxis()
axs.plot()
```
